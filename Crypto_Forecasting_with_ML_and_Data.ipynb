```python
import numpy as np
from qiskit.circuit.library import z_feature_map, real_amplitudes
from qiskit_machine_learning.algorithms import VQC
from qiskit_algorithms.optimizers import COBYLA
from qiskit.primitives import StatevectorSampler

# 1. Toy crypto daily returns (price change %)
# X is the input data, y is the label (1 for Up, 0 for Down)
X = np.array([[0.02], [-0.01], [0.03], [-0.02], [0.01], [-0.03]])
y = np.array([1, 0, 1, 0, 1, 0])

# 2. Define the Quantum Circuit "Brain"
# Using the modern function-based approach to avoid deprecation warnings
# We use 1 qubit because our data has 1 feature (price change)
feature_map = z_feature_map(1, reps=2)
ansatz = real_amplitudes(1, reps=2)

# 3. Initialize the VQC
# We use StatevectorSampler as the modern primitive for simulation
vqc = VQC(
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=COBYLA(maxiter=100),
    sampler=StatevectorSampler()
)

print("Quantum Classifier initialized and ready for training.")
```

```python
# 1. Train the Quantum Model
print("Training in progress...")
vqc.fit(X, y)
print("Quantum model training completed.")

# 2. Predict on new crypto market data (+1.5% return)
new_data = np.array([[0.015]])
prediction = vqc.predict(new_data)

# 3. Output the Trading Signal
print("\n" + "="*30)
print("   QUANTUM TRADING SIGNAL")
print("="*30)
print(f"Input Return: {new_data[0][0]*100}%")

# FIX: We check the prediction value directly.
# If it's an array, we take the first element; if not, we use it directly.
final_pred = prediction[0] if prediction.ndim > 0 else prediction

if final_pred == 1:
    print("Prediction: ðŸ“ˆ PRICE GOES UP")
    print("Action: BUY")
else:
    print("Prediction: ðŸ“‰ PRICE GOES DOWN")
    print("Action: SELL/HOLD")
print("="*30)
```
