```python
from math import gcd
import random

# Number to factor
N = 15

while True:
    # pick a random 'a'
    a = random.randint(2, N-1)

    # check gcd first
    g = gcd(a, N)
    if g > 1:
        result = {'factors': [g, N//g], 'successful': True}
        break

    # pretend quantum period finding (for N=15, r=4)
    r = 4
    x = pow(a, r//2, N)

    if x != N-1:
        p = gcd(x-1, N)
        q = gcd(x+1, N)
        if p > 1 and q > 1:
            result = {'factors': [p, q], 'successful': True}
            break

print(result)
```
